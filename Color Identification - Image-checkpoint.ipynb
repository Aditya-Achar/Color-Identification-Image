# 1) Importing all the necessary Libraries

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt      #matplotlib for plotting related instructions
import numpy as np                   #importing numpy as we need to work on n-dimensional arrays
import cv2                           #importing cv2 from OpenCV-Python for image processing
from collections import Counter
from skimage.color import rgb2lab, deltaE_cie76
import os

%matplotlib inline
```

# 2) Taking the image

```python
image = cv2.imread('tree.jpg')
print("The type of image is",type(image))
print("The dimension of the image is",np.shape(image))
plt.imshow(image)
```

Here, the Type of input is shown as numpy.ndarray. This is because when working with OpenCV Python, images are stored in numpy array.

Also the dimensions of the array is got using the np.shape()
which gives the height, width and no. of channels.

OpenCV by default reads the color order in BGR. Thus the image is somewhat looking discoloured.

We convert the image intO RGB format. 

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

        This is the image in RGB format.

# 3) Applying KMeans Algorithm to identify colors in the image

```python
# KMeans expects flattened array as input. So, we need to reshape the image.reshape().
img = image.reshape((image.shape[0] * image.shape[1] , 3))

# n_clusters specifies the number of dominant colors to be identified from the image.
n_clusters = 8

kmeans= KMeans(n_clusters)
kmeans.fit(img)

colors=kmeans.cluster_centers_  
# colors is a numpy array containing the pixel value of 10 most dominant colors.

labels=kmeans.labels_
# Each pixel value of the image is given a particular label. 
# These labels identify each pixels of the image among the 10 dominant colors.

print("The top 10 colors has pixel values as :\n",colors)
print("\nThe labels given to each pixels in the image are:\n",labels)
print("\nThe size of the labels is : ",np.size(labels))

counts=Counter(labels)          
# Counter() gives the count of the pixels with the respective labels.
print(counts)
counts = dict(sorted(counts.items())) 
# sorted() sorts the the counts in ascending order and dict() creates a dictionary.
print("\nThe number of pixels with each labels are :\n ",counts)

# Now plotting the piechart of the dominant colors identified in the image.
print("\nThe most dominant colors are shown in the below pie chart: \n")
fig=plt.figure(figsize = (12, 6))
plt.pie(counts.values(), labels=counts.keys(), colors = np.array(colors/255))
plt.show()       
```
